```python
import matplotlib.pyplot as plt
import numpy as np

# ✅ Fix the list (missing commas before)
BFRB = [
    'Above ear - pull hair',
    'Forehead - scratch',
    'Forehead - pull hairline',
    'Eyebrow - pull hair',
    'Eyelash - pull hair',
    'Neck - pinch skin',
    'Neck - scratch',
    'Cheek - pinch skin'
]

for gesture in BFRB:
    # Get all unique sequence_ids for this gesture
    ids = df_imu[df_imu['gesture'] == gesture]['sequence_id'].unique()
    
    # If fewer than 3, just take all of them
    sample_ids = np.random.choice(ids, size=min(3, len(ids)), replace=False)
    
    for seq_id in sample_ids:
        # Extract data for this sequence
        df_seq = df_imu.loc[df_imu['sequence_id'] == seq_id].sort_values('sequence_counter')

        # Plot
        plt.figure(figsize=(10, 4))
        plt.plot(df_seq['sequence_counter'], df_seq['acc_x'], label='acc_x')
        plt.plot(df_seq['sequence_counter'], df_seq['acc_y'], label='acc_y')
        plt.plot(df_seq['sequence_counter'], df_seq['acc_z'], label='acc_z')

        plt.title(f'{gesture} — {seq_id}')
        plt.xlabel('sequence_counter')
        plt.ylabel('acceleration')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
```
